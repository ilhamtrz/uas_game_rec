```python
import pandas as pd
import re

# Load your dataset
df = pd.read_csv('merged_data.csv')

# Define a function to extract the numeric value from the text
def extract_number(text):
    match = re.search(r'(\d+,\d+)', str(text))
    if match:
        return int(match.group(0).replace(',', ''))
    return 0  # Return 0 if no match is found

# Apply the function to create the 'Number of Reviews' column
df['Number of Reviews'] = df['All Reviews Number'].apply(extract_number)

# Create a new DataFrame with selected columns
new_df = df[['Title', 'Game Description', 'Number of Reviews']]

# Print the new DataFrame
print(new_df)
```

```python
import nltk
import re
import numpy as np
import contractions

stop_words = nltk.corpus.stopwords.words('english')

def normalize_document(doc):
    if not pd.isnull(doc):  # Check if the value is not NaN
        # lower case and remove special characters\whitespaces
        doc = re.sub(r'[^a-zA-Z0-9\s]', '', doc, re.I | re.A)
        doc = doc.lower()
        doc = doc.strip()
        doc = contractions.fix(doc)
        # tokenize document
        tokens = nltk.word_tokenize(doc)
        # filter stopwords out of the document
        filtered_tokens = [token for token in tokens if token not in stop_words]
        # re-create document from filtered tokens
        doc = ' '.join(filtered_tokens)
        return doc
    else:
        return ''  # Return an empty string for NaN values

normalize_corpus = np.vectorize(normalize_document)

norm_corpus = normalize_corpus(list(top_500_df['Game Description']))
len(norm_corpus)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer(ngram_range=(1, 2), min_df=2)
tfidf_matrix = tf.fit_transform(norm_corpus)
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

doc_sim = cosine_similarity(tfidf_matrix)
doc_sim_df = pd.DataFrame(doc_sim)
doc_sim_df.head()
```

```python
def game_recommender(game_title, games=games_list, doc_sims=doc_sim_df):
    # find game id
    game_idx = np.where(games == game_title)[0][0]
    # get game similarities
    game_similarities = doc_sims.iloc[game_idx].values
    # get top 5 similar game IDs
    similar_game_idxs = np.argsort(-game_similarities)[1:6]
    # get top 5 games
    similar_games = games[similar_game_idxs]
    # return the top 5 games
    return similar_games
```
